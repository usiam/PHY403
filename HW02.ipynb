# 1.e)

```python
def throw():
    score = 1
    radius = 1
    x_list = []
    y_list = []
    while True:
        x, y = np.random.random(2) * 2 - 1 
        distance = np.sqrt(x**2 + y**2)

        if distance > radius: # Game over
            break

        score += 1
        radius = np.sqrt(radius**2 - distance**2) #radius resets to chord length
        x_list.append(x)
        y_list.append(y)
    return score, x_list, y_list
```

```python
a = throw()
s, x, y = a
```

```python
ss, xs, ys = [], [], []
```

```python
for i in range(1001):
    s, x, y = throw()
    xs.append(x)
    ys.append(y)
    ss.append(s)
```

```python
xs = np.concatenate(xs, axis = None)
ys = np.concatenate(ys, axis = None)
ss = np.concatenate(ss, axis = None)
```

```python
plt.hist(ss)
plt.title("Histogram of Scores")
plt.xlabel('Scores')
plt.show()
```

# 1.g)

```python
def throw():
    score = 1
    radius = 1
    x_list = []
    y_list = []
    while True:
        r = np.random.uniform(0,1)
        theta = np.random.uniform(0, 2*np.pi)
        x = r * np.cos(theta)
        y = r * np.sin(theta)
        distance = np.sqrt(x**2 + y**2)
        
        if distance > radius: # Game over
            break

        score += 1
        radius = np.sqrt(radius**2 - distance**2) #radius resets to chord length
        x_list.append(x)
        y_list.append(y)
    return score, x_list, y_list
```

```python
b = throw()
s, x, y = b
```

```python
ss, xs, ys = [], [], []
```

```python
for i in range(1001):
    s, x, y = throw()
    xs.append(x)
    ys.append(y)
    ss.append(s)
```

```python
xs = np.concatenate(xs, axis = None)
ys = np.concatenate(ys, axis = None)
ss = np.concatenate(ss, axis = None)
```

```python
mean = np.mean(ss)
stdev = np.std(ss)

print('mean = ' + str(mean) + ', stdev = ' + str(stdev))
print('The mean score increases compared to other method')
```
