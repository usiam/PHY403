```python
import random
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LogNorm
from matplotlib.pyplot import *

font = {'size': 14}

matplotlib.rc('font', **font)
```

## Generating the variables

```python
np.random.seed(521)
U1 = np.random.uniform(size = 10000)
U2 = np.random.uniform(size = 10000)
R = np.sqrt(-2 * np.log(U1))
Theta = 2 * np.pi * U2
X = R * np.cos(Theta)
Y = R * np.sin(Theta)
r = np.sqrt(X**2 + Y**2)
```

## Plotting for questions 7 a b c d

```python
#7a
plt.plot(X, Y, ".", color = 'purple')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Q7a) Scatter plot of y vs x')
plt.rcParams['axes.facecolor'] = 'black'
plt.show()
```

```python
#7b
counts,xbins,ybins,image = plt.hist2d(X,Y,bins=10)
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Q7b) 2D histogram')
cmap=plt.cm.Greys
plt.show()
```

```python
#7c
plt.contour(counts.transpose(),extent=[xbins[0],xbins[-1],ybins[0],ybins[-1]],
    linewidths=3, cmap = plt.cm.rainbow, levels = [1,5,10,25,50,70,80,100])
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()
plt.rcParams['axes.facecolor'] = 'black'
plt.title('Q7c) 2D contour plot')
plt.show()
```

```python
#7d
plt.hist(r, bins = 20, color = 'white')
plt.xlabel('r')
plt.title(r'Q7d) 1D histogram of r = $\sqrt{x^2 + y^2}$')
plt.show()
```
